```python
import numpy as np
import matplotlib.pyplot as plt
f=open("q8-dataset.dat","r")
lines=f.readlines()
npt=len(lines)
pos=np.zeros((npt,2))
i=0
for line in lines:
    pos[i]=[float(x) for x in line.split()]
    i+=1
x=pos[:,0]
y=pos[:,1]
plt.plot(x,y)
plt.show()
print(len(pos),npt)
```

```python
def fnx(x):
    fnx=0.0
    for i in range(npt):
        f=1.0
        for j in range(npt):
            if j!=i :
                f*=(x-pos[:,0][j])/(pos[:,0][i]-pos[:,0][j])
        f*=pos[:,1][i]
        fnx+=f 
    return fnx
```

```python
x=np.linspace(0,4,100)
y=fnx(x)
plt.plot(x,y)
plt.scatter(pos[:,0],pos[:,1])
plt.show()
```
